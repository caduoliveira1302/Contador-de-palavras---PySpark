# Importar biblioteca

```python
from Scweet.scweet import scrape
from Scweet.utils import init_driver, log_in
import pandas as pd
```

# Scrapar palavras especifos no twitter

```python
driver = init_driver(headless=False, show_images=False, proxy=None)

log_in(driver, env="twitter.env")

data = scrape(words=['whey','suplementos','proteina','treino','crossfit'], since="2022-06-15", until="2023-06-15", from_account = None,interval=1, 
              headless=True, display_type="Top", save_images=False, proxy = None, save_dir = 'outputs',
              resume=False, filter_replies=True, proximity=False, driver=driver)
```

```python
data
```

```python
# Converter data para dataframe
df = pd.DataFrame(data)

# Salvar dataframe para csv
df.to_csv('twitter_whey_V3.csv', index=False)
```

# Pr-anlise:

```python
import pandas as pd

# Carregar o arquivo CSV em um DataFrame
dados_twitter = pd.read_csv('twitter_whey_V3.csv')
```

```python
# Exibir as primeiras linhas do DataFrame
display(dados_twitter.head())

# Verificar as colunas
print(dados_twitter.columns)
```

# Mtrica de engajamento
Para determinar se o Twitter  um bom lugar para divulgar seu produto, voc pode calcular mtricas de engajamento, como mdia de likes, retweets e comentrios.

```python
dados_twitter.dtypes
```

```python
# Calcular a mdia de likes e retweets
dados_twitter['Likes'] = pd.to_numeric(dados_twitter['Likes'], errors='coerce')
media_likes = str(dados_twitter['Likes'].mean())
dados_twitter['Retweets'] = pd.to_numeric(dados_twitter['Retweets'], errors='coerce')
media_retweets = dados_twitter['Retweets'].mean()
```

```python
print("Mdia de likes:", media_likes)
print("Mdia de retweets:", media_retweets)
```

# Grficos
Visualizao de dados

**(mais pra frente, podemos usar esse grfico para comparar com outras plataformas)**

```python
import matplotlib.pyplot as plt

plt.bar(['Likes', 'Retweets'], [float(media_likes), float(media_retweets)])
plt.xlabel('Mtricas')
plt.xlabel('Mtricas')
plt.ylabel('Mdia')
plt.title('Engajamento no Twitter')
plt.show()
```
