```python
import pandas as pd
from translate import Translator
translator = Translator(to_lang='pt')
```

```python
df_reddit = pd.read_csv('C:/Users/cedua/CDIA - PUCSP/PROJETO PySpark - SAVINO/Contador-de-palavras---PySpark/data/final_data_socials/Contagem_reddit.csv')
df_reddit.drop(['Unnamed: 0'], axis=1, inplace=True)
df_reddit = df_reddit.rename(columns={"0":"words", "1": "count"})
df_reddit = df_reddit.head(50)
df_reddit
```

```python
def translate_keeping_whey_protein_reddit(df):
    """""
    Translate the words from dataframe, desconsidering "whey" and "protein"
    """""
    df_p1 = df.loc[[0,2]]                           # "whey" e "protein"
    df_p2 = pd.concat([df.iloc[[1]], df[3:]])       # translate this 

    df_p2['words'] = [translator.translate(x) for x in df_p2['words']]
    df_final = pd.concat([df_p1, df_p2]) #joining the translated df with the "whey" and "protein" words
    return df_final


df_reddit = translate_keeping_whey_protein_reddit(df_reddit)
df_reddit
```

```python
df_twitter = pd.read_csv('C:/Users/cedua/CDIA - PUCSP/PROJETO PySpark - SAVINO/Contador-de-palavras---PySpark/data/final_data_socials/Contagem_Twitter.csv')
df_twitter.drop(['Unnamed: 0'], axis=1, inplace=True)
df_twitter = df_twitter.rename(columns={"Palavra":"words", "Frequência": "count"})
df_twitter = df_twitter.sort_values('count', ascending=False)
df_twitter = df_twitter.reset_index()
df_twitter = df_twitter.drop(['index'], axis=1)
df_twitter = df_twitter.head(50)
df_twitter = df_twitter.drop([20,22,31])
df_twitter
```

```python
def translate_keeping_whey_protein_twitter(df):
    """""
    Translate the words from dataframe, desconsidering "whey" and "protein"
    """""
    df_p1 = pd.concat([df_twitter[df_twitter['words']=='whey'], df_twitter[df_twitter['words']=='protein']]) # "whey" e "protein"
    df_p2 = df.drop([1,3])                                                                                   # translate this 

    df_p2['words'] = [translator.translate(x) for x in df_p2['words']]
    df_final = pd.concat([df_p1, df_p2]) #joining the translated df with the "whey" and "protein" words
    return df_final


df_twitter = translate_keeping_whey_protein_twitter(df_twitter)
```
